Import Libraries

```python
import pandas as pd
import plotly.express as px
```

Data Preparation

```python
# Load the Data
df = pd.read_csv('data/ramen-ratings.csv')

# Check For Nullls then Drop
df['Stars'] = pd.to_numeric(df['Stars'], errors='coerce')
df.dropna(subset=['Stars'], inplace=True)
```

Data Visualization

```python
style_stats = df.groupby('Style')['Stars'].agg(['mean', 'count']).reset_index()
style_stats = style_stats.sort_values(by='mean', ascending=False)

fig = px.bar(
    style_stats, 
    x='Style', 
    y='mean',
    color='mean',               # Colors the bars based on the rating value
    text_auto='.2f',            # Shows the rating number on top of the bars
    labels={'mean': 'Average Rating', 'Style': 'Packaging Style'},
    title='Average Ramen Rating by Packaging Style',
    color_continuous_scale='Viridis', # A professional, eye-pleasing color scale
    template='plotly_white'
)

# Optional: Improve the layout
fig.update_layout(
    xaxis_title="Packaging Style",
    yaxis_title="Average Stars (0-5)",
    coloraxis_showscale=False  # Hides the color bar for a cleaner look
)

fig.show()
```
